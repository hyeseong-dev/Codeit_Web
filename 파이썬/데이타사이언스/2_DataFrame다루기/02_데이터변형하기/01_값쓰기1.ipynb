### 만약 새로운 값을 넣고 싶으면 할당 연산자로 새로운 값을 넣어주면됨.

두 가지 조회방법    
첫째. loc[]    
둘째. iloc[] 

```python
iphone_df.loc['iPhone 8','메모리'] = '2.5GB'
# 첫번쨰는 해당 로우, 두번쨰는 컬럼으로 두게 되면 해당 셀의 값 한개 값이 
# 나온다. 
iphone_df
```

```python
iphone_df.iloc[1,2] = '32.5GB'
iphone_df
```

```python
# iPhone 8의 출시 버전을 바꾸어 봅시다. 
iphone_df.iloc[1,3] = 'iOS 10.3'
iphone_df
```

### 리스트로 개수에 맞게 값들을 다 바꿔 버린다. 

```python
iphone_df.loc['iPhone 8'] = [None, None,None,None,None]
iphone_df
```

# column 하나를 통으로 바꾸고 싶은 경우라면

```python
# 컬럼 하나만을 선택하는 방법 기본형으로 수정하는 방법 
iphone_df.loc[:,'디스플레이']  =['4.7 in','5.5 in','4.7 in','5.5 in','5.8 in','6.5 in']
iphone_df.loc[:,'디스플레이']
```

```python
# 또는 그냥 컬럼 이름만 넣어도 된다. 단축형으로 하는 방법 
iphone_df['출시일'] = ['?','?','?','?','?','?']
iphone_df
```

```python
iphone_df.iloc[:,0] = ['4.7 in','5.5 in',
 '4.7 in',
        '5.5 in',
        '5.8 in',
   '6.5 in']

len(iphone_df)
```

```python
# Face ID 칼럼을 모두 No로 바꿉시다. 
iphone_df['Face ID'] = ['NO' if i>= 0 else 'YES' for i in range(len(iphone_df)) ] # len()을 써서 모든 Face ID 모든(len(iphone_df)) 개수만큼 바꾼거임. 
iphone_df
iphone_df['출시일'] = [None for i in range(len(iphone_df)) if i >= 0]
iphone_df
```

```python
# 걍 문자열 하나만 넣으면 일괄적으로 다 처리함. 
iphone_df['Face ID'] = 'YES'
iphone_df
```

```python
iphone_df['디스플레이'] = [i+0.1 for i in range(len(iphone_df))]
iphone_df
```

```python
iphone_df['Face ID'] = 'NO'
iphone_df
```
